### 문제 5

다음 두 DataFrame을 생성하세요.

```python
# df_student
# 이름  학번
# 철수   1
# 영희   2
# 민수   3

# df_score
# 학번  점수
# 1    90
# 2    80
# 4   100
df_student = pd.DataFrame({
    "이름" : ["철수","영희","민수"],
    "학번" : [1,2,3]
})

df_score = pd.DataFrame({
    "학번" : [1,2,4],
    "점수" : [90,50,100],
})

df_student
```

```python
df_score
```

- ① inner join 결과를 출력하세요.

```python
pd.merge(df_student,df_score,how='inner')
```

- ② outer join 결과를 출력하세요.

```python
pd.merge(df_student,df_score,how='outer')
```

- ③ inner와 outer의 차이를 설명하세요.

```python
# inner은 공통된값만 가져오고 공통되지 않은값은 사라지게한다.
# outer은 공통되지 않은 값 포함 전부 가져온다.
```
