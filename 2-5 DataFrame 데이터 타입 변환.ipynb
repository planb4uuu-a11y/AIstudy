## 📊  2-5 DataFrame 데이터 타입 변환


```python
import pandas as pd
import numpy as np
```



|함수|설명|주 사용 목적|
|-----|----------|--------|
|astype()|일반 타입 변환|숫자 ↔ 숫자
|pd.to_numeric()|숫자로 변환|문자열 → 숫자
|pd.to_datetime()	|날짜형 변환	|문자열 → 날짜
|pd.to_timedelta()	|시간 간격 변환	|문자열 → 날짜 간격

```python
df = pd.DataFrame({"A": [1, 2, 3],
        "B": [1.0, 2.0, 3.0],
        "C": ["2026-01-01", "2026-01-02","2026-01-03"],
        "D": ["1 days", "2 days", "3 days"],
        "E": ["1", "2", "3"]})

df
```

### 1️⃣ 데이터 타입 확인
- 모든 컬럼의 데이터 타입 확인

```python
df.dtypes
```

```python
# 전체 데이터 타입과 Null의 여부 확인가능
df.info()
```

### 2️⃣ astype() : 기본 타입 변환

```python
# A컬럼의 타입
df['A'].dtype
```

```python
# 컬럼 A의 데이터타입을 실수로 변경
df['A'].astype('float')
```

```python
df
```

```python
# 컬럼 B의 데이터타입을 정수로 변경
df['B'].astype('int')
```

```python
df.info()
```

-  컬럼 A와 B를 동시에 다른 타입으로 변경

```python
# 두개를 동시에 바꾸고 싶을땐 딕셔너리 형태, 키:값 형식으로 준다.
df_a = df.astype({'A' : float, 'B' : int})
df_a
```

```python
print(df.dtypes)  # 원본 
print("===========")
print(df_a.dtypes)
```

### 3️⃣ pd.to_numeric() : 숫자형으로 변환
-  컬럼 E를 숫자로 변경

```python
df['E'].dtypes
```

```python
# 숫자로 변환
df_a['E'] = pd.to_numeric(df['E'])
df_a.dtypes
```

### 4️⃣ pd.to_datetime() : 날짜형으로 변환
- 컬럼 C를 datetime으로 변경

```python
df['C']
```

```python
pd.to_datetime(df['C'])

# us = microseconds
```

```python
df_a['C'] = pd.to_datetime(df['C'])
df_a.dtypes
```

### 5️⃣ pd.to_timedelta() : 시간 간격 변환
-  컬럼 D를 timedelta로 변경

```python
df['D']
```

```python
pd.to_timedelta(df["D"])
```

```python
df_a["D"] = pd.to_timedelta(df["D"])
```

```python
print(df.dtypes)  # 원본 
print("===========")
print(df_a.dtypes)
```
