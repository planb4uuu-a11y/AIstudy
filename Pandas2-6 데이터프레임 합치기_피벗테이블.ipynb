## 2-6  Pandas 데이터프레임 합치기 & 피벗테이블

Pandas에서 여러 DataFrame을 합치는 방법은 크게 2가지입니다.

- `pd.concat()` → 단순 연결
- `merge()` → 공통 컬럼 기준 JOIN

```python
import pandas as pd
```

```python
df1 = pd.DataFrame({
    "이름": ["철수", "영희"],
    "점수": [90, 80]
})

df2 = pd.DataFrame({
    "이름": ["민수", "지영"],
    "점수": [85, 95]
})
```

```python
df1
```

```python
df2
```

### 📌 1-1) pd.concat() : 행 기준 합치기 (axis=0)

👉 위아래로 붙이기
👉 기본값 `axis=0`

```python
result = pd.concat([df1, df2], axis=0)
result
```

```python
# 인덱스 재정렬
result = pd.concat([df1, df2], axis=0, ignore_index=True)
result
```

### 📌 1-2) pd.concat() : 열 기준 합치기 (axis=1)

```python
df3 = pd.DataFrame({
    "학년":[1,2]
})
df3
```

```python
# df1, df3 합치기(좌, 우로 붙이기)
result = pd.concat([df1, df3], axis=1)
result
```

```python
result = pd.concat([df1, df2], axis=1, ignore_index=True)
result
```

### 📌 1-3) merge() : 공통 컬럼 기준 JOIN

- 👉 SQL JOIN과 동일
- how 옵션 – inner, left, right, outer

```python
member_df = pd.DataFrame({
                '고객번호': [1001, 1002, 1003, 1004, 1005],
                '이름': ['철수', '영희', '혜진', '도윤', '신우']
            })

member_df
```

```python
transaction_df = pd.DataFrame({
                    '고객번호': [1001, 1001, 1002, 1005, 1006],
                    '금액': [1000, 2000, 1500, 500, 700]
                })
transaction_df
```

##### ✔ inner join
- how = 'inner'
- 공통 데이터만

```python
pd.merge(member_df, transaction_df, how='inner')
```

```python
#left
pd.merge(member_df, transaction_df, how = 'left')

#member_df가 기준이므로 1006번 고객은 나오지 않는다
```

```python
#right
pd.merge(member_df, transaction_df, how = 'right')

#transaction_df를 기준으로 하므로 1003, 1004출력되지 않음
#1006번은 고객번호등록이 되지 않았으므로 NaN으로 처리됨
```

```python
#outer
pd.merge(member_df, transaction_df, how = 'outer')

# ✔ 전체 포함, 없는 값은 NaN
```

## 2️⃣ 피봇 테이블 (Pivot Table)

- 데이터 열 중 두 개의 열을 각각 행 인덱스, 열 인덱스로 사용하여 데이터를 조회하여 펼쳐놓은 것을 말한다.
- pivot 메서드
   - 첫번째 인수 : 행 인덱스로 사용할 열 이름
   - 두번째 인수 : 열 인덱스로 사용할 열 이름
   - 마지막 인수 : 데이터로 사용할 열 이름
   - 만약 주어진 데이터가 존재하지 않으면 해당 칸에 NaN 값을 넣는다.
- df.pivot(index, columns, values)

#### 📌 예제 데이터
- 각 도시의 연도별 인구

```python
df = pd.DataFrame({
    "도시": ["서울", "서울", "서울", "부산", "부산", "부산", "인천", "인천"],
    "연도": ["2015", "2010", "2005", "2015", "2010", "2005", "2015", "2010"],
    "인구": [9904312, 9631482, 9762546, 3448737, 3393191, 3512547, 2890451, 2632030],
    "지역": ["수도권", "수도권", "수도권", "경상권", "경상권", "경상권", "수도권", "수도권"]
})

df
```

- 어떤 도시의 어떤 시점의 인구 파악하기 위해 pivot 테이블로 변경
    - 행 인덱스 : "연도", 열 인덱스 : "도시",  데이터 이름 : "인구"

```python
df.pivot(index="연도", columns="도시", values="인구")
```

```python
df2 = df.pivot(index="연도", columns="도시", values="인구")
df2.fillna(0)
```

```python
df.pivot(index="연도", columns="도시", values="인구").fillna(0)
```

### 3️⃣ pivot_table() 사용하여 결측치 처리

📌 pivot과 pivot_table 차이

|구분	|pivot	|pivot_table|
|---------|-------------|-----------------|
|중복 데이터	|❌ 에러 발생	|✅ 집계 가능|
|집계 함수	|없음	|있음 (mean, sum 등)|
|결측치 처리|	NaN 그대로|	fill_value 사용 가능|
|실무 사용도	|낮음	|⭐⭐⭐⭐⭐|

```python
df1 = pd.DataFrame({
    "이름": ["철수", "철수", "영희", "영희", "영희"],
    "과목": ["수학", "영어", "수학", "영어", "수학"],
    "점수": [90, 85, 80, 95, 70]
})

print(df1)

#👉 영희는 수학 점수가 2개 (중복 존재)
```

```python
# 중복데이터
# df1.pivot(index="이름", columns="과목", values="점수")

#ValueError: Index contains duplicate entries, cannot reshape
```

```python
pd.pivot_table(
    df1,
    index="이름",
    columns="과목",
    values="점수"
)

##🔹 기본 집계 함수 = mean
#👉 영희 수학 점수 (80, 70) → 평균 75
```

```python
df1.pivot_table(
    index="이름",
    columns="과목",
    values="점수"
)
```

### 📌 결측치 처리 (fill_value 사용)
    - 👉 데이터가 없으면 NaN

```python
df
```

```python
df.pivot_table(
    index="연도",
    columns="도시",
    values="인구"
)
```

```python
#fill-value - 결측치 대체
df.pivot_table(
    index="연도",
    columns="도시",
    values="인구",
    aggfunc="sum",
    fill_value=0
)
```

```python
df.pivot_table(
    index="연도",
    columns="도시",
    values="인구",
    aggfunc="sum",
    fill_value=0
)
```
