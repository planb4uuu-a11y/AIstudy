### 문제 1

KMeans 모델을 생성하고 군집을 **3개**로 설정하시오.


```python
import numpy as np
fruits=np.load('./data/fruits_300.npy')
fruits
```

```python
from sklearn.cluster import KMeans
km=KMeans(n_clusters=3, random_state=42, n_init="auto") 
```

```python
fruits_2d=fruits.reshape(-1,100*100)
fruits_2d
```

```python
from sklearn.cluster import KMeans

km = KMeans(n_clusters=3, random_state=42)

km.fit(fruits_2d)   # 🔥 이거 꼭 먼저!

centers = km.cluster_centers_
```

```python
import numpy as np
import matplotlib.pyplot as plt

centers = km.cluster_centers_


centers = centers.reshape(-1, 100, 100)

fig, axs = plt.subplots(1, 3, figsize=(15,5))

for i in range(3):
    axs[i].imshow(centers[i], cmap='gray_r')
    axs[i].set_title(f'center {i}')
    axs[i].axis('off')

plt.show()
```

### 문제 5

K값을 **2~8까지 바꿔 inertia 그래프**를 그리시오.

```python
#여러개 k값 테스트

inertia=[]
for k in range(2,8):
    km=KMeans(n_clusters=k,random_state=42,n_init='auto')
    km.fit(fruits_2d)
    inertia.append(km.inertia_)

plt.plot(range(2,8),inertia)
plt.xlabel('K')
plt.ylabel('inertia')
plt.show()
```
