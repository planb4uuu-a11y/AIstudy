```python
import numpy as np

perch_length = np.array(
    [8.4, 13.7, 15.0, 16.2, 17.4, 18.0, 18.7, 19.0, 19.6, 20.0, 
     21.0, 21.0, 21.0, 21.3, 22.0, 22.0, 22.0, 22.0, 22.0, 22.5, 
     22.5, 22.7, 23.0, 23.5, 24.0, 24.0, 24.6, 25.0, 25.6, 26.5, 
     27.3, 27.5, 27.5, 27.5, 28.0, 28.7, 30.0, 32.8, 34.5, 35.0, 
     36.5, 36.0, 37.0, 37.0, 39.0, 39.0, 39.0, 40.0, 40.0, 40.0, 
     40.0, 42.0, 43.0, 43.0, 43.5, 44.0]
     )
perch_weight = np.array(
    [5.9, 32.0, 40.0, 51.5, 70.0, 100.0, 78.0, 80.0, 85.0, 85.0, 
     110.0, 115.0, 125.0, 130.0, 120.0, 120.0, 130.0, 135.0, 110.0, 
     130.0, 150.0, 145.0, 150.0, 170.0, 225.0, 145.0, 188.0, 180.0, 
     197.0, 218.0, 300.0, 260.0, 265.0, 250.0, 250.0, 300.0, 320.0, 
     514.0, 556.0, 840.0, 685.0, 700.0, 700.0, 690.0, 900.0, 650.0, 
     820.0, 850.0, 900.0, 1015.0, 820.0, 1100.0, 1000.0, 1100.0, 
     1000.0, 1000.0]
     )
```

```python
#훈련데이터와 데스트 데이터로 나눕니다.
#훈련세트와 데트스 세트를 2차원 배열로 바꿉니다.
from sklearn.model_selection import train_test_split
train_input, test_input, train_target, test_target=train_test_split(perch_length,perch_weight, random_state=42)
```

```python
train_input, test_input, train_target, test_target
```

```python
train_input=train_input.reshape(-1,1)
test_input=test_input.reshape(-1,1)
```

```python
train_input, test_input
```

```python
#이웃의 개수를 3으로 설정합니다.
#K-최근점 이웃 회귀모델을 훈련합니다.
from sklearn.neighbors import KNeighborsRegressor
knr=KNeighborsRegressor(n_neighbors=3)
# # knr.n_neighbors=3
# knr=KNeighborsRegressor()
knr.fit(train_input,train_target)
```

```python
knr.score(test_input,test_target)
```

```python
knr.predict([[50]])
```

```python
test_prediction2=knr.predict([[100]])
test_prediction2
```

```python
import matplotlib.pyplot as plt

distances, indexes = knr.kneighbors([[50]])
plt.scatter(train_input, train_target)
plt.scatter(train_input[indexes],train_target[indexes],marker='s')
plt.scatter(50,1033,marker="^")

```

# 선형회귀 모델 구하기
- 다른 알고리즘

```python
from sklearn.linear_model  import LinearRegression
lr=LinearRegression()
lr.fit(train_input,train_target)
```

```python
lr.predict([[100]])
```

# 기울기와 절편 확인


```python
#LinearRegression클래스가 찾은 a와 b는 lr객체의 coef_와 intercept_속성ㅔ 저장되어 있다.

lr.coef_,lr.intercept_
#농어의 무게= a*농어의 길이+b
```

```python
# 훈련 세트 산점도
plt.scatter(train_input, train_target)

# 15~50 구간의 회귀선
plt.plot([15, 50],
         [15*lr.coef_[0] + lr.intercept_,
          50*lr.coef_[0] + lr.intercept_])

# 50cm 농어 예측값
plt.scatter(50, lr.predict([[50]])[0], marker='^')

plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
#훈련세트와 테스트 세트에 대한 점수 확인
lr.score(train_input,train_target)
lr.score(test_input,test_target)
```

```python
#10cm농어에 대한 예측, 마이너스 몸무게나 나옴. 문제가 있음.
lr.predict([[10]])
```

# 그래서 등장 -> 다항 회귀

$ y=ax^2+bx+c$

```python
#2차 방정식의 그래프를 그릴려면 길이를 제곱한 항이 훈련세트에 추가되어야 한다.
```

```python
train_poly=np.column_stack((train_input**2, train_input))
```

```python
test_poly=np.column_stack((test_input**2, test_input))
```

```python
train_poly.shape, test_poly.shape, train_input.shape
```

```python
lr=LinearRegression()
lr.fit(train_poly,train_target)
```

```python
lr.score(test_poly,test_target)
```

```python
lr.predict([[50**2,50]])
```

```python
lr.predict([[10**2,10]])
```

```python
lr.coef_,lr.intercept_
```

```python
1.014*50**2+(-21.55*50)+116
```

```python
#산점도를 그려보자


#구간별 직선을 그리기 위해 15에서 49까지 정수 배열을 만듭니다. 
point=np.arange(15,49)

#훈련세트의 산점도를 그립니다. 
plt.scatter(train_input, train_target)
#15에서 49까지 2차 방정식 그래프를 그립니다. 
plt.plot(point, 1.014*point**2+(-21.55*point)+116 )

#50cm 농어 데이터 
plt.scatter([50],[1573.5],marker="^")

plt.scatter([50],[1573.5],marker="^")
plt.scatter(50,lr.predict([[50**2,50]]),marker="^")
plt.scatter(10,lr.predict([[10**2,10]]),marker="^")
plt.show()
```

```python
tr=lr.score(train_poly,train_target)
te=lr.score(test_poly,test_target)
tr,te
```
