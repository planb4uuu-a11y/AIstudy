```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1,2,3,4,5]
y = [2,4,6,8,10]
plt.plot(x,y, color="red", ls='--')
plt.title("Linear Graph")
plt.xlabel("x valeue")
plt.ylabel('y value')
plt.grid(True,color='gray')
# - 제목: "Linear Graph"
# - x축 이름: "X Value"
# - y축 이름: "Y Value"
# - 그리드 적용
# - 선 색상은 빨강
# - 선 스타일은 점선
# - 그외 조건은 출력형태 참조
```

```python
x = [1,2,3,4,5]
y1 = [1,4,9,16,25]
y2 = [1,2,3,4,5]

plt.plot(x,y1,label="y = x^2",color="blue")
plt.plot(x,y2,label="y = x",color="green")
plt.legend(loc="best")
plt.grid(True,color="gray")
plt.xlabel("x")
plt.ylabel("y")



# - 각각 범례 표시
# - y1은 파란색, y2는 초록색
# - legend 위치는 왼쪽 위
# - x축 이름: "y = x^2"
# - y축 이름: "y = x"
# - 그리드 적용
# - 그외 조건은 출력형태 참조

# plt.figure(figsize=(6,2))
# x=np.arange(0,10)
# y=x**2
# z=x+10

# plt.plot(x,y,label="y=x^2", color='gray', linewidth=5, ls=':', marker='s',ms=10, mec='b',mfc='r')
# plt.plot(z,label="z", color='gray', linewidth=1, ls='--', marker='s',ms=1, mec='b',mfc='g')

# plt.title('Sample Graph')
# plt.xlabel("Time")
# plt.ylabel("Value")
# # plt.legend(loc="best")
# plt.legend(loc="lower right")
# plt.grid(True,color='gray')
# plt.show()
```
