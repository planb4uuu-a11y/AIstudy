## ✔ DataFrame이란?

- 2차원 표 형태
- **index + columns + values**

```python
#딕셔너리-키:값형식으로 만듬
#키는 열의 이름이 됨, 값은 values가 됨, 인덱스는 자동 만들어짐

data={"name":["lee","park","kim","hwang","choi"],
      "score":[100,90,30,40,90],
      "grade":["a","b","c","d"]}
data, type(data)

```

```python
data={"name":["lee","park","kim","hwang","choi"],
      "score":[100,90,30,40,90],
      "grade":["a","b","c","d","f"]}

df=pd.DataFrame(data, columns=["name","grade","score"])
df
```

```python
df.index, df.columns, df.values
#2차원구조임을 알 수 있다.
```

```python
type(df.values)
```

```python
df.shape, df.ndim, df.size
#여기에서 사이즈는 총 데이터 갯수-> 5*3=15
```

# ✔ DataFrame의 index 설정

```python
data={"name":["lee","park","kim","hwang","choi"],
      "score":[100,90,30,40,90],
      "grade":["a","b","c","d","f"]}

df=pd.DataFrame(data, index=["lee","park","kim","hwang","choi"])
df=pd.DataFrame(data, colums=["name","grade","score"])
df


```

```python
df=pd.DataFrame(data)
df
```

```python
#인덱스 재 설정할 수 있다.
df.index=['lee','park','kim','hwang','choi']
df
```

```python
df.drop('name', axis=1), df.drop('score',axis=1)
```
