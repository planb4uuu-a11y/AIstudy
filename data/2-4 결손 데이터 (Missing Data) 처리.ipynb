## 2-4 결손 데이터 (Missing Data) 처리

```python
import pandas as pd
import numpy as np
```

```python
data = {'name':['Lee', 'Hwang', 'Kim', 'Choi'],
      'score':[100, 95, 80, 85],
      'grade':['A', 'A', 'B', np.nan]}
df = pd.DataFrame(data)
df
```

### NaN 값을 가진 컬럼 추가

```python
# point열에 NaN추가
df['point'] = np.nan
df
```

```python
df.isna()
# → 주어진 값이 NaN인지 확인
```

```python
df.isnull()
```

### 2️⃣ fillna()
- → NaN 값을 지정한 값으로 채움

```python
df
```

```python
type(df)
```

```python
# df.fillna(value=0, inplace=True)
df

# fillna는 모든 컬럼에 같은 값을 넣는다.
# 문자열 컬럼에 숫자를 넣으면 TypeError 발생 가능.
```

```python
df.dtypes
```

```python
df.select_dtypes(include='number').columns
```

```python
num_cols = df.select_dtypes(include='number').columns
df[num_cols] = df[num_cols].fillna(0)
```

```python
df
```

```python
df['point2'] = [50, 100, 70, np.nan]
df
```

```python
# - point2 컬럼의 NaN에 평균값으로 채우기
df.fillna(value=df['point2'].mean())
```

```python
df['grade'].dtype

#StringDtype → Pandas에서 제공하는 문자열 전용 dtype
```

```python
df
```

```python
df['grade'] = df['grade'].fillna('0')
df

# ✔ dtype 유지 (string)
```

### 3) df.dropna()
- → NaN이 포함된 행(또는 열)을 삭제

```python
df
```

```python
# df['point2'] = df['point2'].replace(0, np.nan)
# 0을 NaN으로 치환
```

```python
df.dropna(how='any')
```

```python
df

#dropna는 뷰이다, 원본데이터가 변하지 않는다.
```

```python
df.loc[4, :] = [np.nan, np.nan, np.nan, np.nan, np.nan]
df
```

```python
#all, 모두가 결측치 일때
df.dropna(how='all')
```
