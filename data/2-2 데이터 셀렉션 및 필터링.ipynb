# 2️⃣ loc[] — 명칭(Label) 기반 인덱싱 
-df.loc[행이름, 열이름]

```python
df
```

```python
df.loc[0]
```

```python
#여러행 검색 행을 꼭찝어서 이름을 쓰면 됨
df.loc[[0,1,4]]
```

```python
#여러행 검색 범위
#만약 index가 문자인 경우도 정렬된 순서로 범위 슬라이싱이 가능하다. 사전순, sort되어 있어야 한다.
df.loc[0:2]
```

```python
df.loc[0:2,['score','grade']]
df.loc[[0,2],['score','grade']]
```
