```python
df1=pd.DataFrame({
    "이름":["철수","영희2"],
    "점수":[90,80]
})

df2=pd.DataFrame({
    "이름":["민수","지영"],
    "점수":[85,95]
})
```

```python
with pd.ExcelWriter("result.xlsx",engine="openpyxl")as writer:
    df1.to_excel(writer, sheet_name='1반', index=False)
    df2.to_excel(writer, sheet_name='2반',index=False)
```

# 시트추가 저장(기존파일에 추가)

```python
with pd.ExcelWriter("result.xlsx", engine="openpyxl",mode='a') as writer:
    df.to_excel(writer,sheet_name="3반",index=False)
```

여러 시트를 딕셔너리로 저장하는 방법

```python
dfs={
    "1반": df1,
    "2반": df2
}
```

```python

with pd.ExcelWriter('result.xlsx') as writer:
    for sheet, data in dfs.items():
        data.to_excel(writer, sheet_name=sheet, index=False)
```
