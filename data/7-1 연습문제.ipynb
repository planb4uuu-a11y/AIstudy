## 문제 6 (코딩)

epochs를 5 → 10으로 바꿔 정확도를 비교하시오.

```python
from tensorflow import keras
from sklearn.model_selection import train_test_split

# 데이터 로드
(train_input, train_target), (test_input, test_target) = keras.datasets.fashion_mnist.load_data()

# 정규화
train_scaled = train_input / 255.0
train_scaled = train_scaled.reshape(-1, 28*28)

test_scaled = test_input / 255.0
test_scaled = test_scaled.reshape(-1, 28*28)

train_scaled, val_scaled, train_target, val_target = train_test_split(
    train_scaled, train_target, test_size=0.2, random_state=42)

# 모델 생성
dense = keras.layers.Dense(10, activation='softmax', input_shape=(784,))
model = keras.Sequential([dense])

# 컴파일
model.compile(
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# epochs=5
model.fit(train_scaled, train_target, epochs=5, verbose=0)
print("epochs=5 평가")
val_loss, val_acc = model.evaluate(val_scaled, val_target)
test_loss, test_acc  = model.evaluate(test_scaled, test_target)

print(f"검증셋 -> loss: {val_loss:.4f}, accuracy: {val_acc:.4f}")
print(f"테스트셋 -> loss: {test_loss:.4f}, accuracy: {test_acc:.4f}")
```

```python
# 모델 생성
dense = keras.layers.Dense(10, activation='softmax', input_shape=(784,))
model = keras.Sequential([dense])

# 컴파일
model.compile(
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# epochs=5
model.fit(train_scaled, train_target, epochs=5, verbose=0)
print("epochs=10 평가")
val_loss, val_acc = model.evaluate(val_scaled, val_target)
test_loss, test_acc  = model.evaluate(test_scaled, test_target)

print(f"검증셋 -> loss: {val_loss:.4f}, accuracy: {val_acc:.4f}")
print(f"테스트셋 -> loss: {test_loss:.4f}, accuracy: {test_acc:.4f}")
```
