```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind

data1 = np.random.normal(50,10,100)
data2 = np.random.normal(60,15,100)

stat, p = ttest_ind(data1, data2)

plt.boxplot([data1, data2])
plt.xticks([1,2], ["class A", "class B"])
plt.title("Comparison of two groups")

# ---------------------------
# ⭐ 유의성 판단
# ---------------------------
if p < 0.001:
    sig = "***"
elif p < 0.01:
    sig = "**"
elif p < 0.05:
    sig = "*"
else:
    sig = "ns"

# ---------------------------
# ⭐ 선 위치 설정
# ---------------------------
y = max(data1.max(), data2.max()) + 5
h = 2  # 선 높이

# 선 그리기 ( ┐───┌ 모양 )
plt.plot([1, 1, 2, 2], [y, y+h, y+h, y], lw=1.5)

# 별표 표시
plt.text(1.5, y+h, sig, ha='center', va='bottom', fontsize=15)
plt.ylim(0, 130) 
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind

# 재현성
np.random.seed(42)

data1 = np.random.normal(50,10,100)
data2 = np.random.normal(60,15,100)

stat, p = ttest_ind(data1, data2)

# 유의성 기준
if p < 0.001:
    sig = "***"
elif p < 0.01:
    sig = "**"
elif p < 0.05:
    sig = "*"
else:
    sig = "ns"

plt.figure(figsize=(6,6))

# -------------------
# Boxplot
# -------------------
box = plt.boxplot([data1, data2],
                  patch_artist=True,
                  widths=0.45,
                  showfliers=False)

colors = ["#6FA8DC", "#F6B26B"]

for patch, color in zip(box['boxes'], colors):
    patch.set_facecolor(color)
    patch.set_alpha(0.35)
    patch.set_edgecolor("black")
    patch.set_linewidth(1.2)

for median in box['medians']:
    median.set_color("black")
    median.set_linewidth(2)

# -------------------
# Raw data 점 추가 (jitter)
# -------------------
x1 = np.random.normal(1, 0.04, size=len(data1))
x2 = np.random.normal(2, 0.04, size=len(data2))

plt.scatter(x1, data1, color=colors[0], alpha=0.25, s=20)
plt.scatter(x2, data2, color=colors[1], alpha=0.25, s=20)

# -------------------
# 평균 강조
# -------------------
plt.scatter([1,2], [np.mean(data1), np.mean(data2)],
            color="black", s=80, zorder=3)

# -------------------
# 유의성 선
# -------------------
y = max(data1.max(), data2.max()) + 6
h = 2

plt.plot([1, 1, 2, 2], [y, y+h, y+h, y],
         lw=1.5, color="black")

plt.text(1.5, y+h+0.5, sig,
         ha='center', fontsize=18, weight='bold')

# -------------------
# 스타일 정리
# -------------------
plt.xticks([1,2], ["Class A", "Class B"], fontsize=12)
plt.ylabel("Score", fontsize=12)
plt.title("Group Comparison", fontsize=15, weight="bold")

plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

plt.grid(axis='y', linestyle='--', alpha=0.2)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# 재현성
np.random.seed(42)
data = np.random.normal(50, 10, 1000)

mean = np.mean(data)
std = np.std(data)
n = len(data)

plt.figure(figsize=(6.5, 5))

# 히스토그램 (확률밀도)
plt.hist(data, bins=30, density=True,
         edgecolor="black", linewidth=0.8, alpha=0.5)

# 이론 정규분포 곡선
x = np.linspace(min(data), max(data), 300)
plt.plot(x, norm.pdf(x, mean, std), linewidth=2)

# 평균 및 ±1SD
plt.axvline(mean, linestyle="--", linewidth=2)
plt.axvline(mean + std, linestyle=":", linewidth=1.5)
plt.axvline(mean - std, linestyle=":", linewidth=1.5)

# 텍스트 정보 (좌상단 배치)
plt.text(0.02, 0.95,
         f"n = {n}\nMean = {mean:.2f}\nSD = {std:.2f}",
         transform=plt.gca().transAxes,
         verticalalignment='top')

# 라벨 및 스타일
plt.title("Distribution of Scores", fontsize=13, weight="bold")
plt.xlabel("Score", fontsize=11)
plt.ylabel("Density", fontsize=11)

plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde

# 재현성
np.random.seed(42)

data = np.random.normal(50, 10, 1000)

plt.figure(figsize=(7,5))

# 히스토그램
counts, bins, patches = plt.hist(
    data,
    bins=25,
    density=True,           # 확률밀도 형태
    alpha=0.4,
    edgecolor="black",
    linewidth=0.8
)

# KDE 곡선 추가
kde = gaussian_kde(data)
x = np.linspace(min(data), max(data), 300)
plt.plot(x, kde(x), linewidth=2)

# 평균선 추가
mean = np.mean(data)
plt.axvline(mean, linestyle="--", linewidth=2)
plt.text(mean, max(kde(x))*0.9,
         f"Mean = {mean:.2f}",
         rotation=90,
         va='top')

# 스타일 정리
plt.title("Distribution of Data", fontsize=14, weight="bold")
plt.xlabel("Value", fontsize=12)
plt.ylabel("Density", fontsize=12)

plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

plt.grid(axis='y', linestyle='--', alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

np.random.seed(42)

data1 = np.random.normal(50, 10, 1000)
data2 = np.random.normal(60, 10, 1000)

mean1, std1 = np.mean(data1), np.std(data1)
mean2, std2 = np.mean(data2), np.std(data2)

plt.figure(figsize=(7,5))

# 히스토그램 (density)
plt.hist(data1, bins=30, density=True,
         alpha=0.4, edgecolor='black',
         linewidth=0.7, label="Group 1")

plt.hist(data2, bins=30, density=True,
         alpha=0.4, edgecolor='black',
         linewidth=0.7, label="Group 2")

# 정규분포 곡선
x = np.linspace(min(data1.min(), data2.min()),
                max(data1.max(), data2.max()), 400)

plt.plot(x, norm.pdf(x, mean1, std1), linewidth=2)
plt.plot(x, norm.pdf(x, mean2, std2), linewidth=2)

# 평균선
plt.axvline(mean1, linestyle="--", linewidth=1.8)
plt.axvline(mean2, linestyle="--", linewidth=1.8)

# 통계 정보 텍스트 (좌상단)
plt.text(0.02, 0.95,
         f"Group 1: n={len(data1)}, Mean={mean1:.2f}, SD={std1:.2f}\n"
         f"Group 2: n={len(data2)}, Mean={mean2:.2f}, SD={std2:.2f}",
         transform=plt.gca().transAxes,
         verticalalignment='top')

plt.title("Distribution Comparison Between Groups",
          fontsize=13, weight="bold")
plt.xlabel("Score", fontsize=11)
plt.ylabel("Density", fontsize=11)

plt.legend(frameon=False)

# 미니멀 스타일
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

plt.tight_layout()
plt.show()
```
