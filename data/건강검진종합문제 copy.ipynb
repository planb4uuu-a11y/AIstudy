```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [10, 20]
}

grid = GridSearchCV(
    RandomForestClassifier(random_state=42),
    param_grid,
    cv=3,
    n_jobs=1
)

grid.fit(X_train, y_train)

print('best_params', grid.best_params_)
print('best_score', grid.best_score_)
```

```python
#이름과 함께 출력
import pandas as pd
importances=pd.Series(
    grid.best_estimator_.feature_importances_,
    index=X.columns
)
importances.sort_values(ascending=False).head(40)
```

```python
#더미할때 분할되어 있는 데이터는 다 더해줘야 함. 

# 연령대코드(5세단위)_7',
#        '연령대코드(5세단위)_8', '연령대코드(5세단위)_9', '연령대코드(5세단위)_10', '연령대코드(5세단위)_11',
#        '연령대코드(5세단위)_12', '연령대코드(5세단위)_13', '연령대코드(5세단위)_14', '연령대코드(5세단위)_15',
#        '연령대코드(5세단위)_16', '연령대코드(5세단위)_17', '연령대코드(5세단위)_18',
```

```python
#더미 변수들을 그룹으로 합쳐서 중요도 계산
#원핫인고딩된 변수는 importance를 합쳐서 해석하는 것이 일반적이다.
import pandas as pd

# feature importance 가져오기
importances = pd.Series(
    grid.best_estimator_.feature_importances_,
    index=X.columns
)

# 그룹별 중요도 계산
group_importance = {}

for col in importances.index:
    base = col.split('_')[0]  # 원래 변수 이름 추출
    group_importance[base] = group_importance.get(base, 0) + importances[col]

# 결과를 Series로 변환 후 정렬
group_importance = pd.Series(group_importance).sort_values(ascending=False)

# 상위 40개 출력
print(group_importance.head(40))
#.get(key,default)
# key가 있으면- 그 값을 가져옴
```

```python
# permutation importance
특정 변수를 섞었을때 모델 성능이 얼마나 떨어지는지 보고 중요도를 계산
```
